```python
import os, random, sys, time
from urllib.parse import urlparse
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
browser = webdriver.Safari(executable_path='/usr/bin/safaridriver')
browser.maximize_window()
```

```python
browser.get('https://www.linkedin.com/company/10445991/admin/analytics/followers/')
```

```python
file = open('config.txt')
lines = file.readlines()
username = lines[0]
password = lines[1] 

elementID = browser.find_element_by_id('username')
elementID.send_keys(username)

elementID = browser.find_element_by_id('password')
elementID.send_keys(password)

elementID.submit()

time.sleep(5)

elementID = browser.find_element_by_class_name('btn__primary--large')
elementID.click()
```

```python
SCROLL_PAUSE_TIME = 5

#Get scrooll height
last_height = browser.execute_script("return document.body.scrollHeight")

for i in range(3):
    #scrooll down to bottom
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    
    # wait to load page
    time.sleep(SCROLL_PAUSE_TIME)
    
    #calculate new scroll height and compare to last scroll
    new_height = browser.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
src = browser.page_source
soup = BeautifulSoup(src, 'lxml')
soup
```

```python
pop_up_window = soup.find('div', {'class':'full-width'})
pop_up_window
```

```python
js = "'arguments[0].scrollTop = arguments[0].scrollTop + arguments[0].offsetHeight;'",pop_up_window
while True:
    browser.execute_script(js)
    time.sleep(1)  
browser.quit()
```
