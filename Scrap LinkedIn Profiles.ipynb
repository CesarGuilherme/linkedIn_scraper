```python
import os, random, sys, time
from urllib.parse import urlparse
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import pandas as pd
import string
```

```python
#Fazer loing no linkedIn
browser = webdriver.Safari(executable_path='/usr/bin/safaridriver')
browser.maximize_window()

browser.get('https://www.linkedin.com/uas/login')

file = open('config.txt')

lines = file.readlines()

username = lines[0]
password = lines[1] 

elementID = browser.find_element_by_id('username')
elementID.send_keys(username)

elementID = browser.find_element_by_id('password')
elementID.send_keys(password)

elementID.submit()

time.sleep(1)

elementID = browser.find_element_by_class_name('btn__primary--large')
elementID.click()
```

```python
url_file = open('LinkedIn_url_profiles.txt', "r")

url_lines = url_file.readlines()

info = {"url":[],"nome":[],"job":[], "local":[],"empresa":[],"conex√µes":[]}

for line in url_lines:
    value = line.split()
    for i in range(len(value)):
        
        browser.get(value[i])
#         print(i)
        
#         SCROLL_PAUSE_TIME = 2

#         #Get scrooll height
#         last_height = browser.execute_script("return document.body.scrollHeight")

#         for i in range(3):
#             #scrooll down to bottom
#             browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")

#             # wait to load page
#             time.sleep(SCROLL_PAUSE_TIME)

#             #calculate new scroll height and compare to last scroll
#             new_height = browser.execute_script("return document.body.scrollHeight")
#             if new_height == last_height:
#                 break
#             last_height = new_height
        time.sleep(4)
    
        src = browser.page_source
        soup = BeautifulSoup(src)

        name_div = soup.find('div', {'class': 'ph5'})
        # follow_div = soup.find('div', {'class': 'pv-top-card--list'})
        
        if name_div:
            
            name = name_div.find('h1').text
            #print("Nome:",name)

            job = name_div.find('div', {'class':'text-body-medium'}).text.strip()
            #print("Job:",job)

            location = name_div.find('span', {'class':'break-words'}).text.strip()
            #print("Location:",location)

            if not (name_div.find_all('div', {'class':'inline-show-more-text--is-collapsed'})):
                company = ""
            else:
                company = name_div.find('div', {'class':'inline-show-more-text--is-collapsed'}).text.strip()
            #print("Company:",company)

            if not (name_div.find('span', {'class':'t-bold'})):
                connections = ""
            else:
                connections = name_div.find('span', {'class':'t-bold'}).text
            #print("Conections:", connections)

            info["url"].append(line)
            info["nome"].append(name)
            info["job"].append(job)
            info["local"].append(location)
            info["empresa"].append(company)
            info["conex√µes"].append(connections)
        else:
            print("USU√ÅRIO N√ÉO EXISTE!")

print("ACABOU!!!")
```
